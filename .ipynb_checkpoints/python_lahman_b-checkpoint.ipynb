```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
engine = create_engine("postgres+psycopg2://postgres:postgres@localhost:5432/lahman_baseball")
```

```python
# use the connection to run a query using pandas!
batting_df = pd.read_sql("SELECT * FROM batting;", con=engine)
batting_df.head()
```

```python
pschool= pd.read_sql("SELECT * FROM collegeplaying;", con=engine)
pschool.head()
```

```python
field= pd.read_sql("SELECT * FROM fielding;", con=engine)
field.head()
```

```python
salaries= pd.read_sql("SELECT * FROM salaries;", con=engine)
salaries.head()
```

```python
salaries.pop('lgid')
salaries.pop('teamid')
salaries.pop('yearid')
```

```python
people= pd.read_sql("SELECT * FROM people;", con=engine)
people.head()
```

```python
people.pop('birthyear')
people.pop('birthmonth')
people.pop('birthcountry')
people.pop('birthstate')
people.pop('birthcity')
people.pop('birthday')
people.pop('deathday')
people.pop('retroid')
people.pop('bbrefid')
people.pop('deathyear')
people.pop('deathmonth')
people.pop('deathcountry')
people.pop('deathstate')
people.pop('deathcity')
people.pop('bats')
people.pop('throws')
people.pop('debut')
people.pop('finalgame')
people.pop('weight')
people.pop('height')
```

```python
people
```

```python
vandyplay=pschool.loc[pschool['schoolid']=='vandy']
vandyplay
```

```python
vandyplayname = pd.merge(vandyplay,people,on=['playerid'],how='inner')
vandyplayname
```

```python
vand_sal_play = pd.merge(vandyplayname,salaries,on=['playerid'],how='inner')
vand_sal_play
```

```python
vand_sal_play=vand_sal_play.sort_values(by='salary', ascending=False)
vand_sal_play
```
