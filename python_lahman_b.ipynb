```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
engine = create_engine("postgres+psycopg2://postgres:postgres@localhost:5432/lahman_baseball")
```

```python
# use the connection to run a query using pandas!
batting_df = pd.read_sql("SELECT * FROM batting;", con=engine)
batting_df.head()
```

```python
pschool= pd.read_sql("SELECT * FROM collegeplaying;", con=engine)
pschool.head()
```

```python
people= pd.read_sql("SELECT * FROM people;", con=engine)
people.head()
```

```python
people.pop('birthyear')
people.pop('birthmonth')
people.pop('birthcountry')
people.pop('birthstate')
people.pop('birthcity')
people.pop('birthday')
people.pop('deathday')
people.pop('retroid')
people.pop('bbrefid')
```

```python
vandyplay=pschool.loc[pschool['schoolid']=='vandy']
vandyplay
```

```python
vandyplayname = pd.merge(vandyplay,people,on=['playerid'],how='outer')
vandyplayname
```
